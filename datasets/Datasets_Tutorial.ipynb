```python
from datasets import load_dataset

dataset = load_dataset("bnsapa/road-detection")
```

```python
images = dataset.filter(lambda x: x["label"] == 0)
segments = dataset.filter(lambda x: x["label"] == 2)
lane = dataset.filter(lambda x: x["label"] == 1)
```

```python
from datasets import Dataset, Image
```

```python
l = [i['image'] for i in segments['train']]
m = [i['image'] for i in lane['train']]
o = [i['image'] for i in images['train']]
```

```python
train = Dataset.from_dict({'image': o, 'segment': l, 'lane':m})
```

```python
train
```

```python
l = [i['image'] for i in segments['test']]
m = [i['image'] for i in lane['test']]
o = [i['image'] for i in images['test']]
```

```python
test = Dataset.from_dict({'image': o, 'segment': l, 'lane':m})
```

```python
test
```

```python
l = [i['image'] for i in segments['validation']]
m = [i['image'] for i in lane['validation']]
o = [i['image'] for i in images['validation']]
```

```python
validation = Dataset.from_dict({'image': o, 'segment': l, 'lane':m})
```

```python
validation
```

```python
from datasets import DatasetDict
```

```python
data = DatasetDict({
    'train':train,
    'test': test,
    'validation':validation
})
```

```python
data
```

```python
data.push_to_hub("bnsapa/road-detection")
```
